```python
import PyPDF2
import os
# Path to the file on Google Drive
folder_path = '/content/drive/My Drive/PDFS'

documents = []

# Loop through all files in the directory
for filename in os.listdir(folder_path):
    # Check if the file is a PDF
    if filename.endswith('.pdf'):
        # Construct the full path to the PDF file
        pdf_path = os.path.join(folder_path, filename)
        # Open the PDF file
        with open(pdf_path, 'rb') as file:
            # Create a PDF reader object
            pdf_reader = PyPDF2.PdfReader(file)
            # Get the total number of pages
            total_pages = len(pdf_reader.pages)
            print(f'Processing {filename} with {total_pages} pages')
            # Extract text from all pages, just to view it and confirm correctness
            for page_num in range(total_pages):
                page = pdf_reader.pages[page_num]
                text = page.extract_text()
                print(f'--- Page {page_num + 1} ---')
                print(text)
                print('\n\n')
            # Now, Extract text from all pages and store in a single string for EACH PDF
            pdf_text = ""
            for page_num in range(total_pages):
                page = pdf_reader.pages[page_num]
                pdf_text += page.extract_text()
            # Append the extracted text for the current PDF to the 'documents' list
            documents.append(pdf_text)

print(f"Processed {len(documents)} PDF documents.")
```

```python

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

Model comes from https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2

```python
doc_embed = model.encode(documents)
len(doc_embed[0])
```

```python
prompt = "Where can I learn Gen AI in Singapore?"
prompt_embed = model.encode(prompt)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(np.array([prompt_embed]), doc_embed)
top_index = np.argmax(sim)
top_doc = documents[top_index]
print(f"Prompt: {prompt}\nTop response: {top_doc}")
```

```python
sorted = np.argsort(sim[0])[::-1]

ranking = [(documents[i], sim[0][i]) for i in sorted]
ranking
```

https://www.sbert.net/examples/applications/cross-encoder/README.html


```python
from sentence_transformers import CrossEncoder
cross_encoder = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
#cross_encoder = CrossEncoder('cross-encoder/mmarco-mMiniLMv2-L12-H384-v1')
```

```python
pairs = []
for doc in documents:
    pairs.append([prompt, doc])
```

```python
scores = cross_encoder.predict(pairs)

print(f"Prompt: {prompt}\n")
for doc, score in zip(documents, scores):
    print(f"Document: {doc}\nScore: {score}\n")
```

```python
top_index_cross = np.argmax(scores)
top_doc_cross = documents[top_index_cross]
print(f"Prompt: {prompt}\nTop response: {top_doc_cross}")
```
